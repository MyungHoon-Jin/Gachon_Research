```python
df["Income group"].replace({'Upper middle income':'middle', 'High income':'high', 'Lower middle income':'low',
       'Low income':'low'}, inplace=True)
df["Region"].replace({'Europe & Central Asia':'ECA&NA', 'East Asia & Pacific':'EAP',
       'Latin America & Caribbean':'LAC', 'Middle East & North Africa':'MENA',
       'South Asia':'SA', 'Sub-Saharan Africa':'SA', 'North America':"ECA&NA"}, inplace=True)
```

```python
import convergence
```

```python
data = df[df.columns[1:]]
ca_total = convergence.ConvergenceAnalysis(data)
ca_high = convergence.ConvergenceAnalysis(data[data["Income group"]=="high"])
ca_middle = convergence.ConvergenceAnalysis(data[data["Income group"]=="middle"])
ca_low = convergence.ConvergenceAnalysis(data[data["Income group"]=="low"])
ca_eap = convergence.ConvergenceAnalysis(data[data["Region"]=="EAP"])
ca_ecana = convergence.ConvergenceAnalysis(data[data["Region"]=="ECA&NA"])
ca_lac = convergence.ConvergenceAnalysis(data[data["Region"]=="LAC"])
ca_mena = convergence.ConvergenceAnalysis(data[data["Region"]=="MENA"])
ca_sa = convergence.ConvergenceAnalysis(data[data["Region"]=="SA"])
```

```python
ca_total.t_test()
ca_high.t_test()
ca_middle.t_test()
ca_low.t_test()
ca_eap.t_test()
ca_ecana.t_test()
ca_lac.t_test()
ca_mena.t_test()
ca_sa.t_test()
```

```python
ca_total.chi_square_test()
ca_high.chi_square_test()
ca_middle.chi_square_test()
ca_low.chi_square_test()
ca_eap.chi_square_test()
ca_ecana.chi_square_test()
ca_lac.chi_square_test()
ca_mena.chi_square_test()
ca_sa.chi_square_test()
```
