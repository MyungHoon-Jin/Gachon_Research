```python
mean = pd.DataFrame()
mean = pd.concat((mean, CA.mean))
mean.columns = ['Total']
for income in CA.income_mean.keys():
    mean = pd.concat((mean, CA.income_mean[income]), axis=1)
for region in CA.region_mean.keys():
    mean = pd.concat((mean, CA.region_mean[region]), axis=1)
mean.columns = ['Total', 'High income', 'Low income', 'Lower middle income', 'Upper middle income',
                'East Asia & Pacific', 'Europe & Central Asia, North America', 'Latin America & Caribbean', 'Middle East & North Africa', 'South Asia', 'Sub-Saharan Africa']
mean
```

```python
sigma = pd.DataFrame()
sigma = pd.concat((sigma, CA.sigma))
sigma.columns = ['Total']
for income in CA.income_sigma.keys():
    sigma = pd.concat((sigma, CA.income_sigma[income]), axis=1)
for region in CA.region_sigma.keys():
    sigma = pd.concat((sigma, CA.region_sigma[region]), axis=1)
sigma.columns = ['Total', 'High income', 'Low income', 'Lower middle income', 'Upper middle income',
                'East Asia & Pacific', 'Europe & Central Asia, North America', 'Latin America & Caribbean', 'Middle East & North Africa', 'South Asia', 'Sub-Saharan Africa']
sigma
```

```python
gamma = pd.DataFrame()
gamma = pd.concat((gamma, CA.gamma))
gamma.columns = ['Total']
for income in CA.income_sigma.keys():
    gamma = pd.concat((gamma, CA.income_gamma[income]), axis=1)
for region in CA.region_sigma.keys():
    gamma = pd.concat((gamma, CA.region_gamma[region]), axis=1)
gamma.columns = ['Total', 'High income', 'Low income', 'Lower middle income', 'Upper middle income',
                'East Asia & Pacific', 'Europe & Central Asia, North America', 'Latin America & Caribbean', 'Middle East & North Africa', 'South Asia', 'Sub-Saharan Africa']
gamma
```

```python
## XlsxWriter 엔진으로 Pandas writer 객체 만들기
writer = pd.ExcelWriter('sanitation_raw_20190617.xlsx')
 
## DataFrame을 xlsx에 쓰기
sigma.to_excel(writer, sheet_name='sigma')
gamma.to_excel(writer, sheet_name='gamma')
mean.to_excel(writer, sheet_name='mean')

## Pandas writer 객체 닫기
writer.close()
```
